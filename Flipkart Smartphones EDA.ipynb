```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('flipkart_smartphones_2022_03_26_2333_IST.csv')
data.head(3)
```

```python
data.dropna(inplace=True)
data.isnull().any().any()
```

```python
data.nunique()
```

```python
data.brand.unique()
```

```python
redmi_smartphone = data[data['brand'] == 'REDMI']
redmi_smartphone.head()
```

```python
redmi_smartphone[['selling_price','original_price']]
```

```python
redmi_smartphone.title.value_counts()
```

```python
redmi_smartphone.index = range(len(redmi_smartphone.index))
redmi_smartphone.head()
```

```python
redmi_smartphone.tail()
```

```python
import warnings
warnings.filterwarnings('ignore')
new = redmi_smartphone.ram.str.split(' ',n=2,expand=True)
redmi_smartphone['ram_val'] = new[0]
redmi_smartphone.head(3)
```

```python
redmi_smartphone.head(3)
```

### Average Rating of Redmi SmartPhones

```python
plt.figure(figsize=(15,5))
sns.barplot(x=redmi_smartphone['title'],y=redmi_smartphone['avg_rating'])
plt.title('Avg Rating of Redmi SmartPhones ')
plt.xlabel('Phones')
plt.ylabel('Avg Rating')
plt.ylim(ymin=3,ymax=4.7)
plt.xticks(rotation=90)
plt.show()
```

### Rating count of Each Redmi Phone

```python
plt.figure(figsize=(20,8))       
sns.barplot(x=redmi_smartphone['title'],y=redmi_smartphone['ratings_count'])
plt.xticks(rotation=90)
plt.title('Rating Count of Each Model')
plt.xlabel('Phones')
plt.ylabel('Rating Count')
plt.show()
```

```python
redmi_smartphone['ram_val'] = redmi_smartphone['ram_val'].astype(int)
```

### Redmi Phone vs Ram Value

```python
plt.figure(figsize=(20,8))       
sns.barplot(x=redmi_smartphone['title'],y=redmi_smartphone['ram_val'])
plt.xticks(rotation=90)
plt.title('Redmi Title VS Ram Value')
plt.xlabel('Phones')
plt.ylim(ymin=1)
plt.ylabel('RAM Value')
plt.show()
```

### Redmi Phone vs Average Rating

```python
plt.figure(figsize=(20,8))       
sns.barplot(x=redmi_smartphone['title'],y=redmi_smartphone['avg_rating'])
plt.xticks(rotation=90)
plt.title('Redmi Title VS Avg Rating')
plt.xlabel('Phones')
plt.show()
```

### Redmi Phone vs Its Ratings count

```python
plt.figure(figsize=(20,8))       
sns.barplot(x=redmi_smartphone['title'],y=redmi_smartphone['ratings_count'])
plt.xticks(rotation=90)
plt.xlabel('Ratings Count')
plt.ylabel('Phones')
plt.show()
```

### Redmi Phone vs Its Selling Price

```python
plt.figure(figsize=(20,8))       
sns.barplot(x=redmi_smartphone['title'],y=redmi_smartphone['selling_price'])
plt.xticks(rotation=90)
plt.xlabel('Phones')
plt.ylabel('Selling Price')
plt.ylim(ymin=8000)
plt.show()
```
